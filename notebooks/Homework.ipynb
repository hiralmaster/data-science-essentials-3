```python
#Create `income_dict`, a dictionary that uses the numerical codes in the income bucket column as keys 
#and the matching descriptions 
income_dict = {1: 'Under $1', 2: 'Between 1 and $10,000', 3: 'Between 10,000 and $25,000', 4: 'Between 25,000 and $50,000', 
               5: 'Between 50,000 and $75,000', 6: 'Between 75,000 and $100,000', 7: 'Between 100,000 and $200,000', 
              8: '$200,000 or more'}
print (income_dict)
```

# Question 7



```python
#Pass the dictionary as an argument to the pandas `replace()` method to change the `income_bucket` column 
#so that it uses descriptive text instead of the numeric code.
income_df['income_bucket'] = income_df['income_bucket'].replace(income_dict)
print(income_df)
income_df.info()
```

```python
#Create a new dataframe `income_county_agg` to show totals for each county. 
#Use the pandas `groupby()` method to group the data by county and get the `sum()` of all numeric columns for that county. 
income_county_agg = income_df.groupby('county').sum()
```
