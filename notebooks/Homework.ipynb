```python
#Build a model that will predict whether the average cost for cancer care 
#is above the average cost or equal to or below the average cost
tn_df.head(-1)
```

```python
tn_df['cancer_cost_income_ratio_above_avg_y'] = tn_df.cancer_cost_income_ratio >  0.431653
tn_df.head()
```

```python
tn_df = pd.get_dummies(tn_df, columns = ['urban'], drop_first = True)
```

```python
tn_df.head()
```

```python
#log reg model
x = tn_df[['urban_Urban']]
y = tn_df['cancer_cost_income_ratio_above_avg_y']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

from cm import plot_confusion_matrix

import warnings

warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x, y)
```

```python
fig, ax = plt.subplots(figsize = (3,4))
sns.barplot(x = logistic_model.coef_[0], y = list(x.columns), ax = ax, edgecolor = 'black')
plt.title('Logistic Regression Coefficients')
```
